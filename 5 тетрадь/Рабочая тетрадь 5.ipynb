```python
import math

class Trigonometry:
    
    @staticmethod
    def cos(angle):
        print(f"    {angle} ")
        return math.cos(angle)
    
    @staticmethod
    def sin(angle):
        print(f"    {angle} ")
        return math.sin(angle)
    
    @staticmethod
    def tan(angle):
        print(f"    {angle} ")
        return math.tan(angle)
    
    @staticmethod
    def asin(value):
        print(f"    {value}")
        return math.asin(value)
    
    @staticmethod
    def acos(value):
        print(f"    {value}")
        return math.acos(value)
    
    @staticmethod
    def atan(value):
        print(f"    {value}")
        return math.atan(value)
    
    @staticmethod
    def to_radians(degrees):
        print(f"  {degrees}   ")
        return math.radians(degrees)


angle = Trigonometry.to_radians(45)
print(Trigonometry.cos(angle))

value = Trigonometry.sin(angle)
print(Trigonometry.asin(value))
```

```python
class Tree:
    
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data

    #    
    def print_tree(self):
        if self.left:
            self.left.print_tree()
        print(self.data)
        if self.right:
            self.right.print_tree()
        
    def insert(self, value):
        if value < self.data:
            if self.left is None:
                self.left = Tree(value)
            else:
                self.left.insert(value)
        elif value > self.data:
            if self.right is None:
                self.right = Tree(value)
            else:
                self.right.insert(value)

root = Tree(50)
root.insert(30)
root.insert(20)
root.insert(40)
root.insert(70)
root.insert(60)
root.insert(80)

root.print_tree()
```

```python
import numpy as np
from sklearn import tree
from sklearn.metrics import classification_report, confusion_matrix

#  
X = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
target = np.array([0, 0, 0, 1, 1, 1])

#   
clf = tree.DecisionTreeClassifier()

#    
clf.fit(X, target)

#   
tree.plot_tree(clf)
```

```python
#    
y_pred = clf.predict(X)

#       
print(confusion_matrix(y_pred, target))
print(classification_report(y_pred, target))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn import metrics

url = 'https://raw.githubusercontent.com/likarajo/petrol_consumption/master/data/petrol_consumption.csv'
dataframe = pd.read_csv(url)
dataframe.describe().round(3)
```

```python
X = dataframe[['Average_income','Population_Driver_licence(%)','Petrol_Consumption']]
y = dataframe['Petrol_tax']
print(X)
```

```python
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size = 0.2,
    random_state = 0
)
# X_train  y_train    ,  
#   . X_test  y_test    , 
#        , 
#       .
```

```python
regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train)
print('   : ')
plot_tree(regressor)
```

```python
y_pred = regressor.predict(X_test)
print("    y    ,    X_test: ")
print(y_pred)
```

```python
df = pd.DataFrame({"Actual": y_test, "Predicted": y_pred})
print("        : ")
print(df.head())
```

```python
print("  MSE: ", metrics.mean_squared_error(y_test, y_pred).round(3))
print("   MAE: ", metrics.mean_absolute_error(y_test, y_pred).round(3))
print("%   MAE: ", (metrics.mean_absolute_error(y_test, y_pred) / np.average(y) * 1))

# MSE ,        ,
#  MAE        .
#  MAE ,       
```
